### Level 1 - Basic

```sql
-- Show all employees whose salary is NULL
SELECT * 
FROM employeess1
WHERE salary IS NULL;
```

```sql
-- Orders where discount is NOT NULL
SELECT * 
FROM Orders1
WHERE discount IS NOT NULL;
```

```sql
-- Products where category is NULL
SELECT * 
FROM Products1
WHERE category IS NULL;
```

```sql
-- Count employees with NULL manager_id
SELECT COUNT(*) AS null_manager_count
FROM employeess1
WHERE manager_id IS NULL;
```

### LEVEL 2 - ISNULL

```sql
-- Replace NULL salary with 0
SELECT emp_id, name, COALESCE(salary, 0) AS salary
FROM employeess1;
```

```sql
-- Replace NULL bonus with 1000
SELECT emp_id, name, COALESCE(bonus, 1000) AS bonus
FROM employeess1;
```

```sql
-- Replace NULL order amount with 500
SELECT order_id, COALESCE(amount, 500) AS amount
FROM Orders1;
```

```sql
-- Replace NULL stock with 0
SELECT product_id, COALESCE(stock, 0) AS stock
FROM Products1;
```

### LEVEL 3 - COALESCE

```sql
-- Employee earnings: salary, if NULL use bonus
SELECT emp_id, name, COALESCE(salary, bonus) AS earnings
FROM employeess1;
```

```sql
-- First available value: salary → bonus → 0
SELECT emp_id, name, COALESCE(salary, bonus, 0) AS final_income
FROM employeess1;
```

```sql
-- Product price default 1000
SELECT product_id, COALESCE(price, 1000) AS price
FROM Products1;
```

```sql
-- Customer payment: amount → discount → 0
SELECT order_id, COALESCE(amount, discount, 0) AS payment
FROM orders1;
```

### LEVEL 4 - NULLIF

```sql
-- Salary becomes NULL if salary = 0
SELECT emp_id, NULLIF(salary, 0) AS salary
FROM employeess1;
```

```sql
--  Discount becomes NULL if discount = 0
SELECT order_id, NULLIF(discount, 0) AS discount
FROM orders1;
```

```sql
-- Avoid divide by zero
SELECT order_id, amount / NULLIF(discount, 0) AS result
FROM orders1;
```

```sql
-- Replace 'DISC10' with NULL
SELECT order_id, NULLIF(coupon_code, 'DISC10') AS coupon_code
FROM orders1;
```

### LEVEL 5 - REAL-TIME SCENARIOS

```sql
-- Total earnings (handle NULL properly)
SELECT emp_id, COALESCE(salary, 0) + COALESCE(bonus, 0) AS total_earnings
FROM employeess1;
```

```sql
-- Employees where both salary AND bonus are NULL
SELECT *
FROM employeess1
WHERE salary IS NULL AND bonus IS NULL;
```

```sql
-- Products where price is NULL but category is NOT NULL
SELECT *
FROM products1
WHERE price IS NULL AND category IS NOT NULL;
```

```sql
-- Orders where both amount and discount are NULL
SELECT *
FROM orders1
WHERE amount IS NULL AND discount IS NULL;
```

### LEVEL 6 - ADVANCED

```sql
-- Employee income with fallback
SELECT emp_id, COALESCE(salary, bonus, 1000) AS income
FROM employeess1;
```

```sql
-- Replace discount 0 with NULL
SELECT order_id, NULLIF(discount, 0) AS discount
FROM orders1;
```

```sql
-- Final payable amount (handle NULL)
SELECT order_id,COALESCE(amount, 0) - COALESCE(discount, 0) AS final_amount
FROM orders1;
```

```sql
-- Employees where salary is NULL but manager exists
SELECT *
FROM employeess1
WHERE salary IS NULL AND manager_id IS NOT NULL;
```
