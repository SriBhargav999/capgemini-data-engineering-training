```sql
-- extract the numeric characters that appear at the very beginning of the value
SELECT mixed_value, REGEXP_SUBSTR(mixed_value, '^[0-9]+') AS start_numbers
FROM regex_practice;
```

```sql
-- extract the numeric characters that appear at the very end of the value
SELECT mixed_value, REGEXP_SUBSTR(mixed_value, '[0-9]+$') AS end_numbers
FROM regex_practice;
```

```sql
-- extract only the first single character of the value, regardless of whether it is a letter or a number
SELECT mixed_value, REGEXP_SUBSTR(mixed_value, '^.') AS first_char
FROM regex_practice;
```

```sql
-- extract only the last single character of the value
SELECT mixed_value, REGEXP_SUBSTR(mixed_value, '.$') AS last_char
FROM regex_practice;
```

```sql
-- extract exactly two consecutive numeric characters that appear anywhere in the value
SELECT mixed_value,REGEXP_SUBSTR(mixed_value, '[0-9]{2}') AS two_digits
FROM regex_practice;
```

```sql
-- extract exactly one numeric character that appears anywhere in the value
SELECT mixed_value, REGEXP_SUBSTR(mixed_value, '[0-9]') AS one_digit
FROM regex_practice;
```

```sql
-- extract the country code present at the beginning of the phone number
SELECT phone, regexp_extract(phone, '[0-9]+', 0) AS country_code
FROM regex_practice;
```

```sql
-- extract the numeric portion that is present between alphabetic characters
SELECT mixed_value, regexp_extract(mixed_value, '[a-zA-Z]+([0-9]+)[a-zA-Z]+', 1) AS middle_numbers
FROM regex_practice;
```

```sql
-- extract the text that appears before the at-symbol
SELECT email,REGEXP_SUBSTR(email, '^[^@]+') AS username
FROM regex_practice;
```

```sql
-- extract the text that appears after the at-symbol including the domain name
SELECT email, regexp_extract(email, '@(.+)', 1) AS domain_full
FROM regex_practice;
```

```sql
-- extract only the domain name without including the at-symbol
SELECT email, regexp_extract(email, '@([^\\.]+)', 1) AS domain
FROM regex_practice;
```

```sql
-- extract only the text that appears after the last dot in the email address
SELECT email, regexp_extract(email, '\\.([^.]+)$', 1) AS extension
FROM regex_practice;
```

```sql
-- extract only alphabetic characters that appear together as a continuous sequence
SELECT mixed_value,REGEXP_SUBSTR(mixed_value, '[a-zA-Z]+') AS letters
FROM regex_practice;
```

```sql
-- extract only numeric characters that appear together as a continuous sequence
SELECT mixed_value, REGEXP_SUBSTR(mixed_value, '[0-9]+') AS numbers
FROM regex_practice;
```

```sql
-- extract exactly the first three characters of the value
SELECT full_text, REGEXP_SUBSTR(full_text, '^.{3}') AS first_three
FROM regex_practice;
```

```sql
-- extract exactly the last two characters of the value
SELECT full_text, REGEXP_SUBSTR(full_text, '.{2}$') AS last_two
FROM regex_practice;
```

```sql
-- extract the employee number portion that appears between the alphabetic prefix and the first underscore
SELECT full_text, regexp_extract(full_text, '[A-Z]+([0-9]+)_', 1) AS emp_number
FROM regex_practice;
```

```sql
-- extract the country code that appears at the end of the value
SELECT full_text,REGEXP_SUBSTR(full_text, '[0-9]+$') AS country_code
FROM regex_practice;
```

```sql
-- extract the alphabetic text that appears between two underscore characters
SELECT full_text, REGEXP_SUBSTR(full_text, '_([A-Z]+)_') AS middle_text
FROM regex_practice;
```

```sql
-- extract the numeric characters that appear immediately after the plus sign
SELECT phone, regexp_extract(phone, '\\+([0-9]+)', 1) AS country_code
FROM regex_practice;
```
