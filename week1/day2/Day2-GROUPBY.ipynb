### GROUP BY with Conditions


```sql
-- total salary for departments where total salary> 100,000
SELECT department, SUM(salary) AS total_salary
FROM employee
GROUP BY department
HAVING total_salary>100000
```

```sql
-- Count the number of employees in departments with more than 2 employees
SELECT department, COUNT(emp_id) as count_of_employees
FROM employee
GROUP BY department
HAVING count_of_employees>2
```

```sql
-- average salary for employees who joined after 2021-01-01
SELECT emp_name, AVG(salary) AS avg_salary
FROM employee
WHERE joining_date>'2021-01-01' 
GROUP BY emp_name
```

```sql
-- departments where the maximum salary is greater than 75,000
SELECT department, MAX(salary) as max_salary
FROM employee
GROUP BY department
HAVING max_salary>75000
```

```sql
-- departments where the total salary is less than 150,000
SELECT department, SUM(salary) as total_salary
FROM employee
GROUP BY department
HAVING total_salary<150000
```

### GROUP BY with HAVING

```sql
-- total number of employees grouped by department with more than 1
SELECT department, COUNT(emp_id) AS Total_employees
FROM employee
GROUP BY department 
HAVING Total_employees>1
```

```sql
-- total salary of each department who are >125000
SELECT department, SUM(salary) AS total_salary
FROM employee
GROUP BY department 
HAVING total_salary>12500
```

```sql
-- Count the number of employees in each department with > 2 emp
SELECT department, COUNT(emp_id) AS Employee_count
FROM employee
GROUP BY department 
HAVING Employee_count >2
```

```sql
-- average salary in each department >60000
SELECT department , AVG(salary) AS avg_salary
FROM employee
GROUP BY department 
HAVING avg_salary> 60000
```

```sql
-- departments where the sum of salaries is between 100,000 and 200,000
SELECT department, SUM(salary) AS total_salary
FROM employee
GROUP BY department
HAVING total_salary BETWEEN 100000 AND 200000
```

### GROUP BY with JOINs

```sql
-- total sales amount for each employee
SELECT e.emp_name AS employee_name, SUM(s.amount) AS total_amount
FROM employee e 
JOIN sales s ON e.emp_id = s.emp_id
GROUP BY employee_name
```

```sql
-- number of sales made by each employee
SELECT e.emp_name AS employee_name, COUNT(s.sales_id) AS No_Of_Sales
FROM employee e 
JOIN sales s ON e.emp_id=s.emp_id
GROUP BY employee_name
```

```sql
-- total sales amount grouped by product
SELECT product, SUM(amount) AS total_sales_amount
FROM sales
GROUP BY product
```

```sql
-- average sales amount grouped by product
SELECT product, AVG(amount) AS avg_sales_amount
FROM sales
GROUP BY product
```

```sql
-- employees who have made more than 2 sales
SELECT e.emp_name AS employee_name, COUNT(s.sales_id) AS sales_count
FROM employee e 
JOIN sales s ON e.emp_id=s.emp_id
GROUP BY employee_name
HAVING sales_count>2
```

### Advanced GROUP BY

```sql
-- total salary and total sales amount for each employee
SELECT e.emp_name AS employee_name, CAST(SUM(e.salary) AS INT) AS total_salary, CAST(SUM(s.amount) AS INT) AS total_amount
FROM employee e 
JOIN sales s ON e.emp_id=s.emp_id
GROUP BY employee_name
```

```sql
-- Count the number of unique products sold by each employee
SELECT e.emp_name AS employee_name, COUNT(DISTINCT s.product) AS Unique_products
FROM employee e 
JOIN sales s ON e.emp_id=s.emp_id
GROUP BY employee_name
```

```sql
-- highest sales amount made by each employee
SELECT e.emp_name AS employee_name, CAST(MAX(s.amount) AS INT) AS highest_sale_amount
FROM employee e 
JOIN sales s ON e.emp_id=s.emp_id
GROUP BY employee_name
```

```sql
-- total sales amount grouped by product and filtered by products where the total exceeds 50,000
SELECT product, SUM(amount) AS total_sales_amount
FROM sales
GROUP BY product
Having total_sales_amount>50000
```

```sql
-- departments with the highest average sales amount
SELECT e.department, AVG(s.amount) AS avg_sales_amount
FROM employee e 
JOIN sales s ON e.emp_id=s.emp_id
GROUP BY e.department
ORDER BY avg_sales_amount DESC
-- LIMIT 1
```

### Real-World Scenarios

```sql
-- department with the highest total sales amount
SELECT department, CAST(SUM(amount) AS INT) AS total_sales_amount
FROM sales
JOIN employee ON sales.emp_id = employee.emp_id
GROUP BY department
ORDER BY total_sales_amount DESC
LIMIT 1
```

```sql
-- top 3 employees with the highest total sales amount
SELECT e.emp_name AS employee_name, SUM(s.amount) AS total_sales_amount
FROM employee e
JOIN sales s ON e.emp_id=s.emp_id
GROUP BY employee_name
ORDER BY total_sales_amount DESC
LIMIT 3
```

```sql
-- total number of employees and the average salary
SELECT joining_date,  Count(emp_name) AS employee_count, CAST(AVG(salary) AS INT) AS avg_salary
FROM employee e
GROUP BY joining_date
```

```sql
-- employees who have not made any sales
SELECT e.emp_name AS employee_name
FROM employee e 
JOIN sales s ON e.emp_id=s.emp_id
WHERE s.amount IS NULL
```
